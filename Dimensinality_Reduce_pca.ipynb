```python
#importing the modules
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import datasets as dd
```

```python
#load dataset WINE
data = dd.load_wine()
x = pd.DataFrame(data.data)
y = pd.DataFrame(data.target)
data.feature_names,data.target_names
```

```python
x
```

```python
#fitting the data
pca = PCA(n_components=2)
pca.fit(x)
dd = pca.transform(x)
print(dd.shape)
```

```python
#changing the array of features into data frame
xval = pd.DataFrame(dd)
xval.columns=['x0','x1']
xval.head()
xval['x1']
```

```python
#displaying the components
print(pca.components_)
ddf = pd.DataFrame(pca.components_)
ddf
```

```python
#heat map
plt.figure(figsize=(9,5))
sns.heatmap(ddf.corr(),annot=True)
plt.show()
```

```python
#visuilising the scatter plot
plt.scatter(xval['x0'],xval['x1'],c=y)
plt.show()
```

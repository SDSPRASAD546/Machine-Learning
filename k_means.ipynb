```python
#importing modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans as KM
import seaborn as sns
```

```python
#read csv 
Raw_data = pd.read_csv("/diabetes.csv")
```

```python
#make csv to dataframe
data = pd.DataFrame(Raw_data)
```

```python
#printing top 5 values
data.head()
```

```python
#finding the shape
data.shape
```

```python
#finding the null 
data.isnull().sum()
```

```python
#descibing the dataset
data.describe()
```

```python
#plotting the two attribute i.e glucose and age
plt.scatter(data['glucose'].values,data['age'].values)
plt.show()
```

**Performing the KMeans Algorithm**

```python
km = KM(n_clusters=2)
km.fit(data)
lable = km.predict(data)
lable # predicted values
```

```python
#visualising the clustering
plt.scatter(data['glucose'].values,data['age'].values,c=lable)
plt.show()
```
